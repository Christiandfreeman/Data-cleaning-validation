```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df= pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv',encoding = "ISO-8859-1")
df['Journal title']=  df['Journal title'].str.lower()
df.tail()
```

```python
top5= df['Journal title'].value_counts()
top5_names =top5[:5].index.values
df_top5 = df.loc[df['Journal title'].isin(top5_names)]
                
```

```python
df_top5=df_top5.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)':'cost'})
df_top5.head()
```

```python
cost_split=df_top5['cost']
cost_split=cost_split.str.split('£',expand=True)
cost2=cost_split[1]
cost2.sort_values('index',ascending=False)
df_top5['cost']=pd.to_numeric(cost2)
```

```python
df_top5['cost'].dropna()
df_top5.head()
```
